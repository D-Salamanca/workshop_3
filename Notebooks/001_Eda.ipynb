```python

# Renombrar columnas y agregar una columna 'year' para cada dataset
data_2015_renamed = data_2015.rename(columns={
    'Country': 'Country',
    'Happiness Score': 'Happiness Score',
    'Economy (GDP per Capita)': 'Economy (GDP per Capita)',
    'Family': 'Social support',
    'Freedom': 'Freedom',
    'Trust (Government Corruption)': 'Perceptions of corruption',
    'Happiness Rank': 'Happiness Rank',
    'Health (Life Expectancy)': 'Health (Life Expectancy)',
    'Dystopia Residual': 'Dystopia Residual'
}).assign(year=2015)

data_2016_renamed = data_2016.rename(columns={
    'Country': 'Country',
    'Happiness Score': 'Happiness Score',
    'Economy (GDP per Capita)': 'Economy (GDP per Capita)',
    'Family': 'Social support',
    'Freedom': 'Freedom',
    'Trust (Government Corruption)': 'Perceptions of corruption',
    'Happiness Rank': 'Happiness Rank',
    'Health (Life Expectancy)': 'Health (Life Expectancy)',
    'Dystopia Residual': 'Dystopia Residual'
}).assign(year=2016)

data_2017_renamed = data_2017.rename(columns={
    'Country': 'Country',
    'Happiness.Score': 'Happiness Score',
    'Economy..GDP.per.Capita.': 'Economy (GDP per Capita)',
    'Family': 'Social support',
    'Freedom': 'Freedom',
    'Trust..Government.Corruption.': 'Perceptions of corruption',
    'Happiness.Rank': 'Happiness Rank',
    'Health..Life.Expectancy.': 'Health (Life Expectancy)',
    'Dystopia.Residual': 'Dystopia Residual',
    'Whisker.high': 'Upper Confidence Interval',
    'Whisker.low': 'Lower Confidence Interval'
}).assign(year=2017)

data_2018_renamed = data_2018.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'Economy (GDP per Capita)',
    'Social support': 'Social support',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Perceptions of corruption',
    'Overall rank': 'Happiness Rank',
    'Healthy life expectancy': 'Health (Life Expectancy)'
}).assign(year=2018)

data_2019_renamed = data_2019.rename(columns={
    'Country or region': 'Country',
    'Score': 'Happiness Score',
    'GDP per capita': 'Economy (GDP per Capita)',
    'Social support': 'Social support',
    'Freedom to make life choices': 'Freedom',
    'Perceptions of corruption': 'Perceptions of corruption',
    'Overall rank': 'Happiness Rank',
    'Healthy life expectancy': 'Health (Life Expectancy)'
}).assign(year=2019)

# Concatenar todos los datasets con las columnas unificadas
combined_data = pd.concat([data_2015_renamed, data_2016_renamed, data_2017_renamed, data_2018_renamed, data_2019_renamed], ignore_index=True)

# Mostrar las primeras filas del dataset combinado
combined_data.head()
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continents = {
            'AF': 'Africa',
            'AS': 'Asia',
            'EU': 'Europe',
            'NA': 'North America',
            'SA': 'South America',
            'OC': 'Oceania',
            'AN': 'Antarctica'
        }
        return continents[continent_code]
    except:
        return 'Unknown'

# Crear la columna de continente
combined_data['Continent'] = combined_data['Country'].apply(country_to_continent)

# Mostrar algunas filas para confirmar la conversión
print(combined_data[['Country', 'Continent']].head())
```

### Justificación de las Variables Seleccionadas para el Modelo de Felicidad

Para el modelo predictivo del puntaje de felicidad, se seleccionaron las siguientes variables clave:

1. **Economy (GDP per Capita)**: Mide la estabilidad económica y el poder adquisitivo, correlacionándose fuertemente con la felicidad (correlación ≈ 0.79).
2. **Social support**: Representa la red de apoyo social, un factor crucial para el bienestar con una correlación significativa (≈ 0.65).
3. **Health (Life Expectancy)**: Indica la calidad y duración de vida, lo cual impacta directamente en la felicidad (correlación ≈ 0.74).
4. **Freedom**: Refleja la autonomía personal y la capacidad de tomar decisiones, con una correlación moderada (≈ 0.55) con la felicidad.
5. **Perceptions of corruption**: Afecta la confianza en las instituciones, influyendo negativamente en la calidad de vida, aunque con una correlación baja (≈ 0.13).

### Razones para Descartar Otras Variables

- **Happiness Rank** y **Confidence Intervals**: Son valores derivados del `Happiness Score` y no representan características independientes.
- **Dystopia Residual**: Es un valor artificial de referencia y no un factor real que influya en la felicidad.
- **Standard Error**: Representa la variabilidad en la estimación del puntaje y no una característica del país.

Estas variables fueron seleccionadas por su relevancia y correlación directa con el puntaje de felicidad, mientras que las descartadas no aportan valor explicativo adicional.

```python
import numpy as np

selected_data = combined_data[["Continent", "year", "Happiness Score", "Economy (GDP per Capita)", 
                               "Social support", "Health (Life Expectancy)", "Freedom", 
                               "Perceptions of corruption"]]

# Crear interacciones entre variables
selected_data['Economy_SocialSupport'] = selected_data['Economy (GDP per Capita)'] * selected_data['Social support']
selected_data['Health_Freedom'] = selected_data['Health (Life Expectancy)'] * selected_data['Freedom']
selected_data['Economy_Health'] = selected_data['Economy (GDP per Capita)'] * selected_data['Health (Life Expectancy)']

# Aplicar transformaciones no lineales
selected_data['Log_Economy'] = np.log(selected_data['Economy (GDP per Capita)'] + 1)  # Se suma 1 para evitar log(0)
selected_data['Health_Squared'] = selected_data['Health (Life Expectancy)'] ** 2
selected_data['Freedom_Squared'] = selected_data['Freedom'] ** 2

# Ver las primeras filas con las nuevas características
print(selected_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular la matriz de correlación para el conjunto de datos actualizado
correlation_matrix = selected_data.corr()

# Crear el gráfico de calor para la matriz de correlación
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Matriz de Correlación para Variables Seleccionadas y Nuevas Características")
plt.show()
```

### Selección de Variables del Feature Engineering para el Modelo

Después de realizar el Feature Engineering, se seleccionaron solo las variables que mostraron una correlación significativa con el `Happiness Score` y que no redundan con las variables originales:

1. **Economy_SocialSupport**: Esta variable captura el efecto combinado de la economía y el apoyo social, dos factores que por separado ya son influyentes en la felicidad. La combinación de ambos muestra una correlación fuerte con el `Happiness Score` (0.82), lo que sugiere que los países con buena economía y apoyo social tienden a ser más felices.

2. **Economy_Health**: Esta interacción mide el impacto conjunto de la economía y la salud en la felicidad, con una correlación significativa de 0.82 con el `Happiness Score`. Refleja cómo la combinación de recursos económicos y una buena expectativa de vida contribuyen a un mayor bienestar.

3. **Health_Squared**: Elevar al cuadrado la expectativa de vida saludable ayuda a capturar una relación no lineal entre salud y felicidad. La correlación de 0.77 indica que este factor es relevante para el modelo, ya que refleja que la felicidad puede aumentar más en países donde la expectativa de vida es muy alta.

4. **Freedom_Squared**: Aunque `Freedom` por sí sola es significativa, su cuadrado captura un efecto no lineal en la felicidad, con una correlación de 0.56, lo que añade un valor explicativo adicional sin redundar demasiado.

### Variables No Seleccionadas

Algunas variables generadas mediante Feature Engineering fueron descartadas para evitar redundancias y mejorar la simplicidad del modelo:
- **Log_Economy**: Aunque muestra una alta correlación, es prácticamente redundante con `Economy (GDP per Capita)`, por lo que no añade información adicional significativa.
- **Health_Freedom**: Su correlación con la felicidad es más baja (0.69) y, al combinar salud y libertad, no parece aportar tanto valor explicativo como otras interacciones.
- **Freedom_Squared**: Aunque seleccionada para análisis, puede ser eliminada si el modelo resulta en multicolinealidad debido a su alta correlación con `Freedom`.

En conclusión, se seleccionaron solo las variables generadas mediante Feature Engineering que demostraron un impacto claro en la predicción de la felicidad y no duplicaron la información ya proporcionada por las variables originales.
